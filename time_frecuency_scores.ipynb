```python
import matplotlib.pyplot as plt
import numpy as np
import os

from mne import create_info
from mne.time_frequency import AverageTFR


from jupyterthemes.stylefx import set_nb_theme
set_nb_theme('gruvboxd')
```

```python
directory = os.fsencode(os.path.abspath(os.getcwd()))
```

```python
freqs = np.array([  7.        ,  13.33333333,  19.66666667,  26.        ,
        32.33333333,  38.66666667,  45.        ,  51.33333333,
        57.66666667,  64.        ,  70.33333333,  76.66666667,
        83.        ,  89.33333333,  95.66666667, 102.        ,
       108.33333333, 114.66666667, 121.        , 127.33333333,
       133.66666667, 140.        ])
centers = [-0.30000000000000004,
 -0.20000000000000004,
 -0.10000000000000003,
 -2.7755575615628914e-17,
 0.09999999999999998,
 0.19999999999999998,
 0.30000000000000004,
 0.4,
 0.5000000000000001,
 0.6,
 0.7000000000000001,
 0.7999999999999999,
 0.9,
 0.9999999999999999,
 1.0999999999999999,
 1.1999999999999997,
 1.2999999999999998,
 1.4,
 1.5,
 1.6]
```

```python
scores = []
for dr in os.listdir(directory):
    if os.path.isdir(dr) :
        for file in os.listdir(dr):
            if '2s' in str(file) or '0.2s' in str(file) :
                scr = np.load(str(dr).split("'")[1]+'/'+str(file).split("'")[1])
                if scr.shape[1] == 20:
                    scores.append(scr)
```

```python
mean = np.nanmean(scores, axis=0)
```

```python
av_tfr = AverageTFR(create_info(['sfreq'],1024), mean[np.newaxis,:],
                   centers, freqs[1:],1)
```

```python
av_tfr.plot([0], vmin=0.5, title="2s Decoding Scores",
            cmap=plt.cm.Reds)
```
